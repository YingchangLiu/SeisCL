| SeisCL Constants | $\hspace{2cm}$ Description $\hspace{2cm}$ | Default | 
| :--: | :--: | :--: |
| N | $\hspace{7cm}$  Grid size \[ NX, NY, NZ \]  $\hspace{7cm}$ | [250, 150] |
| ND | Flag for dimension. <br><br> 3 : 3D $\hspace{5mm}$ 2 : 2D P-SV $\hspace{5mm}$ 21 : 2D SH \* \* \* $\hspace{5mm}$ 22 : 2D acoustic | 2 |
| dh | Grid spatial spacing | 10 |
| dt | Time step size | 0.0008 |
| NT | Number of time steps | 875 |
| freesurf | Include a free surface <br><br> 0 : no $\hspace{5mm}$ 1 : yes | 0|
| FDORDER | Order of the finite difference <br><br> Possible values : 2,4,6,8,10,12 | 8 |
| MAXRELERROR | Select FD coefficients <br><br> 0 : Taylor $\hspace{5mm}$ 1 : Holberg | 1 |
| L | Number of attenuation mechanism (L=0 elastic) | 0 |
| f0 | Central frequency of the relaxation | 15 |
| FL | Frequencies of the attenuation mechanism | [15] \* \* \* np.array(15)|
| src_pos | Position of each shots. <br><br> Array \[sx sy sz srcid src_type\] x nb sources <br><br> See [Sources and Receivers](Sources&Receivers.ipynb) for more information | - |
| rec_pos | Position of the receivers. <br><br> [gx gy gz srcid recid - - -] x nb receivers <br><br> See [Sources and Receivers](Sources&Receivers.ipynb) for more information | - |
| src | Source signals <br><br> [NT x number of sources] \* \* \*  | - |
| abs_type | Absorbing boundary type: <br><br> 1 : CPML $\hspace{5mm}$ 2: Absorbing layer of Cerjan | 1 |
| VPPML | Vp velocity near CPML boundary | 3500 |
| NPOWER | Exponent used in CPML frame update | 2 |
| FPML | Dominant frequency of the wavefield | 15 |
| K_MAX_CPML | Coeffient involved in CPML (may influence simulation stability) | 2 |
| nab | Width in grid points of the absorbing layer | 16 |
| abpc | Exponential decay for absorbing layer of Cerjan | 6 |
| pref_device_type | Type of processor used: <br><br> 2 : CPU $\hspace{5mm}$ 4: GPU $\hspace{5mm}$ 8: Accelerator | 4 |
| nmax_dev | Maximum number of GPUs | 1 |
| no_use_GPUs | Array of device numbers that should not be used for computation | np.empty( (1,0) )  |
| MPI_NPROC_SHOT | Maximum number of MPI process (nodes) involved in domain decomposition | 1 |
| back_prop_type | Type of gradient calculation: <br><br> 1 : backpropagation (elastic only) $\hspace{5mm}$ 2 : Discrete Fourier transform | 1 $\rightarrow$ backpropagation |
| param_type | Type of parametrization : <br><br> 0 : ($\rho$,$v_p$,$v_s$,$\tau_p$,$\tau_s$) $\hspace{5mm}$ 1 : ($\rho$, M, $\mu$, $\tau_p$, $\tau_s$) $\hspace{5mm}$ 2 : ($\rho$, $I_p$, $I_s$, $\tau_p$, $\tau_s$) | 0 $\rightarrow$  ($\rho$,$v_p$,$v_s$,$\tau_p$,$\tau_s$) |
| gradfreqs | Frequencies of gradient with DFT | - |
| tmax | Maximum time of gradient computation | 0 |
| tmin | Minimum time of gradient computation | 0 |
| scalerms | Scale each modeled and recorded traces according to its rms value, <br> then scale residual by recorded trace rms | 0 |
| scalermsnorm | Scale each modeled and recorded traces according to its rms value, normalized | 0 |
| scaleshot | Scale all of the traces in each shot by the shot total rms value | 0 |
| fmin | Maximum frequency of gradient \* \* \* | 0 |
| fmax | Minimum frequency of gradient \* \* \* | 0 |
| mute | Muting matrix \[5 x number of traces\].  \[t1 t2 t3 t4 flag \] <br><br> t1 to t4 are mute time with cosine tapers <br><br> 0 : keep data in window <br><br> 1 : mute data in window | None |
| weight | NTxnumber of geophones or 1x number of geophones. <br><br> Weight each sample, or trace, according to the value of weight for gradient calculation. | None |
| gradout | Output gradient <br><br> 0 : no <br><br> 1 : yes | 0 |
| Hout | Output approximate Hessian <br><br> 0 : no $\hspace{1cm}$ 1 : yes | 0 |
| gradsrcout | Output source gradient <br><br> 0 : no $\hspace{1cm}$ 1 : yes | 0|
| seisout | Output seismograms <br><br> 1 : output velocities <br><br> 2 : output pressure <br><br> 3 : output stresses <br><br> ??4?? output everything | 2 |
| resout | Output residuals <br><br> 0 : no $\hspace{1cm}$ 1 : yes | 0 |
| rmsout | Output rms value <br><br> 0 : no $\hspace{1cm}$ 1 : yes | 0 |
| movout | Output movie every n frames | 0 |
| restype | Type of cost function <br><br> 0 : raw seismic trace cost function <br><br> 1: Migration | 0 |
| inputres | Input the residuals for gradient computation | 0 |
| FP16 | FP16 computation: <br><br> 0 : FP32 <br><br> 1 : FP32 vector <br><br> 2 : FP16 IO  <br><br> 3 : FP16 IO + COMP | 0 |
